```python
import sys
import os
import os.path
import fnmatch
from lxml import etree
from lxml.etree import tostring
import re
from collections import defaultdict
from collections import OrderedDict
import csv
import copy
```

```python
def get_canon (path_to_csv):
    tmp_dict=defaultdict(list)
    canon_dict=dict()
    with open(path_to_csv, newline='') as csv_file:
        reader = csv.reader(csv_file, delimiter=',')
        for idx,row in enumerate(reader):
            if idx>0:
                if "," in row[0]:
                    name=row[0].split(",")[0]
                    name=name.replace(" ","-")
                    if "non-canon" in row[4]:
                        tmp_dict[name].append("non-canon")
                    else:
                        tmp_dict[name].append("canon")
                else:
                    name=row[0].replace(" ","-")
                    if "non-canon" in row[4]:
                        tmp_dict[name].append("non-canon")
                    else:
                        tmp_dict[name].append("canon")
    
    for k,v in tmp_dict.items():
        nb_non=v.count("non-canon")
        if nb_non>len(v)/2:
            canon_dict[k]="non-canon"
        else:
            canon_dict[k]="canon"
    
    return canon_dict
```

```python
def write_out (path_in, path_out,list_of_files, canon_dict):
    for file in list_of_files:
        file=file.replace("/",":")
        fullPath=path_in+file

        if os.path.isfile(fullPath):
            tree=etree.parse(fullPath)
            
            if tree.findall(".//author"):
                author=tree.findall(".//author")
                name=author[0].attrib["name"]
                if "," in name:
                    name=name.split(",")[0].replace(" ","-")
                else:
                    name=name.replace(" ","-")
                
                if name in canon_dict.keys():
                    if tree.findall(".//profileDesc[@tag='canon']") and "non-canon" in canon_dict[name]:
                        print("rectification en non-canon pour "+name)
                        att_to_change=tree.findall(".//profileDesc[@tag='canon']")[0]
                        att_to_change.attrib['tag']="non-canon"
                        f = open(path_out+file,'w')
                        f.write(etree.tostring(tree, pretty_print=True, encoding = "unicode"))
                        f.close()
                    elif tree.findall(".//profileDesc[@tag='non-canon']") and canon_dict[name] in "canon":
                        print("rectification en canon pour "+name)
                        att_to_change=tree.findall(".//profileDesc[@tag='non-canon']")[0]
                        att_to_change.attrib['tag']="canon"
                        f = open(path_out+file,'w')
                        f.write(etree.tostring(tree, pretty_print=True, encoding = "unicode"))
                        f.close()
```

```python
restreint='/home/odysseus/Bureau/ANR/stats/corpus_restreint/'
romans='/home/odysseus/Bureau/ANR/stats/romans/'
test_folder_restreint='/home/odysseus/Bureau/ANR/stats/test_restreint/'
test_folder_romans='/home/odysseus/Bureau/ANR/stats/test_romans/'
files_restreint=fnmatch.filter(os.listdir(restreint), '*.xml')
files_romans=fnmatch.filter(os.listdir(romans), '*.xml')
canon_dict=get_canon('./canon.csv')
print(canon_dict)

write_out(restreint,test_folder_restreint,files_restreint,canon_dict)
write_out(romans,test_folder_romans,files_romans,canon_dict)
```
